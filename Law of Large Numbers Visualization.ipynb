# Law of Large Numbers Visualization

We will take a look at 4 different situations that will hold all other variable constants and only increase sample sizes. The law of large numbers states that as we increase our sample size our data points will converge towards the center.

For this procedure we will be using Matplotlib and Numpy to generate our number range as well as 

```python
#importing Libraries
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
np.random.seed(15003)
```

Lets begin with our first case wheere our sample size is 10.

```python
#Initializing variables
mean=10
std=2
n=10
#Sampling from a normal distribution
frequency=np.random.normal(mean,std,n)

#Histogram
plt.hist(frequency,bins=30,color="tab:cyan")
plt.title("N= " + str(n))
plt.ylabel("Frequency")
plt.xlabel("Mean= 10")
plt.show()
```

Lets increase our sample size to 100.

```python
#New sample size
n=100
#Sampling from a normal distribution
frequency=np.random.normal(mean,std,n)
#Histogram
plt.hist(frequency,bins=30,color="tab:cyan")
plt.title("N= " + str(n))
plt.ylabel("Frequency")
plt.xlabel("Mean= 10")
plt.show()
```

Lets increase it even more to 10,000.

```python
#New sample size
n=10000
#Sampling from a normal distribution
frequency=np.random.normal(mean,std,n)
#Histogram
plt.hist(frequency,bins=30,color="tab:cyan")
plt.title("N= " + str(n))
plt.ylabel("Frequency")
plt.xlabel("Mean= 10")
plt.show()
```

Lets change our sample size one more time to 1,000,000!

```python
#New sample size
n=1000000
#Sampling from a normal distribution
frequency=np.random.normal(mean,std,n)
#Histogram
plt.hist(frequency,bins=30,color="tab:cyan")
plt.title("N= " + str(n))
plt.ylabel("Frequency")
plt.xlabel("Mean= 10")
plt.show()
```

As we can see it is shown that the law of large numbers holds. As we increased our sample size our points converged towards the mean of 10.
